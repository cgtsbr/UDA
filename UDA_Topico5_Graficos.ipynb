Neste script você verá:
 - Como criar gráficos com matplotlib
 - Como visualizar métricas do DataFrame
 - Como comunicar informações de forma visual


1. Carregando os dados

```python
import pandas as pd

# Supondo uma planilha 'dados_empresa.xlsx'
df = pd.read_excel("dados_empresa.xlsx")

df.head()
```

2. Criando métricas utilizadas nos gráficos

```python
df["Lucro"] = df["Faturamento"] - df["Custos"]
df["Variação (%)"] = df["Faturamento"].pct_change() * 100
```

3. Importando matplotlib

```python
import matplotlib.pyplot as plt
```

4. GRÁFICO 1 — Faturamento Mensal

```python
plt.figure(figsize=(10,5))
plt.plot(df["Mes"], df["Faturamento"])
plt.title("Faturamento Mensal")
plt.xlabel("Mês")
plt.ylabel("Faturamento (R$)")
plt.grid(True)
plt.show()
```

5. GRÁFICO 2 — Lucro por Mês

```python
plt.figure(figsize=(10,5))
plt.bar(df["Mes"], df["Lucro"])
plt.title("Lucro por Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro (R$)")
plt.show()
```

6. GRÁFICO 3 — Faturamento vs Custos

```python
plt.figure(figsize=(10,5))
plt.plot(df["Mes"], df["Faturamento"])
plt.plot(df["Mes"], df["Custos"])
plt.title("Comparação: Faturamento x Custos")
plt.xlabel("Mês")
plt.ylabel("Valores (R$)")
plt.legend(["Faturamento", "Custos"])
plt.grid(True)
plt.show()
```

7. GRÁFICO 4 — Faturamento por Região

```python
# Agrupamento por região
grp_regiao = df.groupby("Região")[["Faturamento", "Lucro"]].sum()

plt.figure(figsize=(10,5))
plt.bar(grp_regiao.index, grp_regiao["Faturamento"])
plt.title("Faturamento Total por Região")
plt.xlabel("Região")
plt.ylabel("Faturamento (R$)")
plt.show()
```

8. GRÁFICO 5 — Percentual de Faturamento por Região

```python
plt.figure(figsize=(8,8))
plt.pie(grp_regiao["Faturamento"], labels=grp_regiao.index, autopct="%1.1f%%")
plt.title("Participação no Faturamento por Região")
plt.show()
```

9. GRÁFICO 6 — Correlação Faturamento x Lucro

```python
plt.figure(figsize=(8,6))
plt.scatter(df["Faturamento"], df["Lucro"])
plt.title("Correlação entre Faturamento e Lucro")
plt.xlabel("Faturamento (R$)")
plt.ylabel("Lucro (R$)")
plt.grid(True)
plt.show()
```

10. GRÁFICO EXTRA

```python
plt.figure(figsize=(10,5))
plt.plot(df["Mes"], df["Variação (%)"])
plt.title("Variação Percentual do Faturamento")
plt.xlabel("Mês")
plt.ylabel("Variação (%)")
plt.axhline(0)
plt.grid(True)
plt.show()
```
